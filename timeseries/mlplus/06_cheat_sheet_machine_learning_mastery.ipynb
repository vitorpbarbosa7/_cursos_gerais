## Autoregression

### Austra

```python
passengers = pd.read_csv('data/AirPassengers.csv', index_col='Month', parse_dates=True)
passengers = passengers['#Passengers']
series = passengers
```

```python
plt.plot(series)
```

```python
from statsmodels.tsa.ar_model import AutoReg
```

```python
plot_pacf(series)
```

```python
# fit model
model = AutoReg(series, lags=[1,2,5,13])
model_fit = model.fit()
```

```python
model_fit.summary()
```
